```python
import json
for filename in folder_content:
    if filename.endswith(r'.geojson'):
        file_path=folder_path+"\\"+filename
        with open(file_path,'r') as f:
            json_data=json.load(f)
```

```python
geo_data=pd.DataFrame.from_dict(json_data,orient='columns')
geo_data
```

```python
geo_dict={'latitude':[],'longitude':[],'neighbourhood':[]}

for f in geo_data.features:
    coordinates=f['geometry']['coordinates']
    neighbourhood=f['properties']['neighbourhood']
    new_coordinates=[]
    new_coordinates.extend(inner for outer in coordinates for middle in outer for inner in middle)
    for c in new_coordinates:
        geo_dict['longitude'].append(c[0])
        geo_dict['latitude'].append(c[1])
        geo_dict['neighbourhood'].append(neighbourhood)
```

```python
neighbourhood_geodata=pd.DataFrame(geo_dict,columns=geo_dict.keys())
neighbourhood_geodata
```

```python
import scipy,shapely
from shapely import Point,Polygon
from scipy.spatial import ConvexHull
poly_dict={}
na_neighbourhood=property_data[property_data.neighbourhood.isnull()]

for key,val in neighb_dict.items():
    stack=np.vstack(val)
    hull=ConvexHull(stack)
    poly=Polygon([stack[vertex] for vertex in hull.vertices])
    poly_dict[key]=poly

for i,r in na_neighbourhood.iterrows():
    coords=Point(r['latitude'],r['longitude'])
    for n,poly in poly_dict.items():
        if poly.contains(coords):
            property_data.at[i,'neighbourhood']=n
            break
    
```
